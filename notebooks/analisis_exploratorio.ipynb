```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/dataset_sintetico.csv")
df.head()
```

```python
df.describe()
```

```python
plt.scatter(df['Temperatura'], df['Ventas'])
plt.title('Relación entre Temperatura y Ventas')
plt.xlabel('Temperatura')
plt.ylabel('Ventas')
plt.show()
```

```python
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Mapa de Correlaciones')
plt.show()
```

```python
Conclusiones del análisis:

La variable que más influye en ventas es Temperatura.

También se nota relación significativa con Promoción y Fin de semana.

El heatmap confirma correlación positiva fuerte entre todas las variables y las ventas.

No se observan valores fuera de lo normal.
```
